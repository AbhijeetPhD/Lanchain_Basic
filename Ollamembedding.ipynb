```python
from langchain_ollama import OllamaEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from dotenv import load_dotenv
import os

```

```python
load_dotenv()
```

### Data Loader

```python
path=r"F:\Generative_AI\Langchain Basic\Lanchain_Basic\Data\speech.txt"
```

```python
loader=TextLoader(path)
```

```python
doc=loader.load()
```

```python
doc
```

### Chunking

```python
chunk=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=3)
```

```python
doc_chunk=chunk.split_documents(doc)
```

```python
doc_chunk
```

### Embedding

```python
embedding=OllamaEmbeddings(model="gemma2:2b")
```

```python
query="what is your name"

embedding_query=embedding.embed_query(query)
```

```python
embedding_query
```

### Vector DB

```python
db=FAISS.from_documents(doc_chunk,embedding)
```

```python
query="what this document is about?"
```

```python
db.similarity_search_with_relevance_scores(query)
```
